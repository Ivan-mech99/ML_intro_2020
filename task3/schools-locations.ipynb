```python
import pandas as pd
import numpy as np
df = pd.read_csv('Rus_schools_final.csv',encoding='1251')
df.head(3)
df.iloc[0:4]
```

```python
smol_list=[]
a=df.shape[0]
for i in range(0,a):
    hold=i
    string1=df.iloc[i]['addr']
    if string1.find('г. Смоленск')!=-1:
        smol_list.append(hold)
        
```

```python
SMOL=pd.DataFrame(columns=['name','struct','addr','lat','lon'])
for i in range(0,len(smol_list)):
    SMOL.loc[i]=[df.iloc[smol_list[i]]['name'],df.iloc[smol_list[i]]['struct'],df.iloc[smol_list[i]]['addr'],df.iloc[smol_list[i]]['lat'],df.iloc[smol_list[i]]['lon']]
SMOL
```

```python
my_ind=0
for i in range(0,len(SMOL)):
    if SMOL.iloc[i]['name'].find("16")!=-1:
        my_ind=i
my_ind
```

```python
SMOL.iloc[30]['name']
```

```python
import folium
map = folium.Map(location=[54.782526, 32.044128],tiles='Stamen Toner',zoom_start = 10.5)
```

```python
for i in range(0,len(SMOL)):
    if i!=my_ind:
       folium.CircleMarker(location=[SMOL.iloc[i]['lat'], SMOL.iloc[i]['lon']],popup=SMOL.iloc[i]['name'].encode('raw_unicode_escape').decode('utf-8'), radius = 9, fill_color='green', color="gray", fill_opacity = 0.9 ).add_to(map)
folium.CircleMarker(location=[SMOL.iloc[30]['lat'], SMOL.iloc[30]['lon']], radius = 18, popup=SMOL.iloc[30]['name'].encode('raw_unicode_escape').decode('utf-8'), fill_color='yellow', color="gray", fill_opacity = 0.9 ).add_to(map)
```
